```
#Нарисуйте график функции
 
import numpy as np
import matplotlib.pyplot as plt


X = np.linspace(-2*np.pi, 3*np.pi, 201)
k = (3.0)
a = (1.0)
b = (2.0)

Y  = k * np.cos(X-a)+b
Y2 = 2*k*np.cos(X-2*a)+2*b   #k,a,b умножены на 2
Y3 = (k-1)*np.cos(X-a-1)+b-1 #k,a,b уменьшены на 1
plt.title('График функции Y = k * np.cos(x-a)+b')
plt.plot(X,  Y)
plt.plot(X, Y2)
plt.plot(X, Y3)
plt.show()

```

```
#Перевод полярных координат в декартовы
#Прошу помочь!!! Не могу понять почему код так странно считает синусы и косинусы. Часть ответов например 0 и 10
#абсолютно правильная, а часть полная ерунда. Наприме 90 и 10. Y считает правильно а Х уже врет.


import math
import matplotlib.pyplot as plt

T = int(input('Введите Полярный угол в градусах  '))
T = math.radians(T)
R = int(input('Полярный радиус                   '))

Y = R * math.sin(T)
X = R * math.cos(T)

print('X', X)
print('Y', Y)

```

```
#График окружности в полярных координатах

import matplotlib.pyplot as plt
import numpy as np

f = np.arange(0., 360., 1.)/180 *np.pi  

plt.polar(f, [0.5]*len(f))  # окружность

plt.show()
```
