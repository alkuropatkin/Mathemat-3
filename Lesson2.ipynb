```
#Окружность 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-2.0, 2.0, 100)
y = np.linspace(-2.0, 2.0, 100)
r = 2
X, Y = np.meshgrid(x,y)
F = X**2 + Y**2 - r**2
plt.title('Окружность')
plt.contour(X,Y,F, 1)
plt.show()
```

```
#Элипс
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1.0, 1.0, 100)
y = np.linspace(-1.0, 1.0, 100)
X, Y = np.meshgrid(x,y)
F = X**2/5 + Y**2/2 - 1
plt.title('Эллипс')
plt.contour(X,Y,F, 0)
plt.show()
```

```
#Плоскость
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

a = float(input('a = '))
b = float(input('b = '))
c = float(input('c = '))
d = float(input('d = '))

xs = np.arange(-5, 5, 0.5)
zs = np.arange(-5, 5, 0.5)

X, Z = np.meshgrid(xs, zs)
Y  = (-d-a*X-c*Z)/b
Y1 = (-a*X-c*Z)/b #праллельная плоскость, проходящая через начало координат

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z,)
ax.plot_wireframe(X, Y1, Z,)
ax.scatter(0, 0, 0, 'z',50,'red')
plt.title('Праллельные плоскости')
plt.show()
```

```
#Поверхность второго порядка.
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

xs = np.arange(-5, 5, 0.5)
ys = np.arange(-5, 5, 0.5)

X, Y = np.meshgrid(xs, ys)
Z=np.sqrt(X**2+Y**2)



fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
plt.show()
```
